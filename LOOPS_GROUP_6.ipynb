```python
# 1. Write a loop that prints the first 5 multiples of 3
print("First 5 multiples of 3:")
for i in range(1, 6):
    print(3 * i)
print()
```

```python
# 2. Given a list of scores, write a loop that counts how many scores are
# greater than 50
print("Count scores greater than 50:")
scores = [45, 67, 32, 89, 56, 41, 72, 98, 34, 87]
count = 0
for score in scores:
    if score > 50:
        count += 1
print(f"Number of scores greater than 50: {count}")
print()
```

```python
# 3. Write a program using a while loop that prints numbers from 10 down to 1
print("Numbers from 10 down to 1:")
num = 10
while num >= 1:
    print(num)
    num -= 1
print()
```

```python
# 4. Given a string, write a loop that prints each character one by one
print("Print each character:")
text = "Group Six"
for char in text:
    print(char)
print()
```

```python
# 5. Write a loop that stops when it encounters the number 7 in a list of numbers
print("Stop when encountering 7:")
numbers = [2, 4, 6, 8, 10, 7, 12]
for num in numbers:
    if num == 7:
        print(f"Found 7! End.")
        break
    print(num)
print()
```

```python
# 6. Write a program that prints numbers from 1 to 10 using a loop
print("Numbers from 1 to 10:")
for i in range(1, 11):
    print(i)
print()
```

```python
# 7. Given a list of names, write a loop that prints each name
print("Print each name:")
names = ["Roua", "Alija", "Chinonyelum", "Goodness", "Mildred", "Bunmi", "Esther", "Onyinye"]
for name in names:
    print(name)
print()
```

```python
# 10. What will be the output of a loop that starts at 1 and increases by 2 until it reaches 10?
print(" Loop starting at 1, increasing by 2 until 10:")
for i in range(1, 11, 2):
    print(i)
print()
print("Output explanation: 1, 3, 5, 7, 9 (stops before exceeding 10)")
```
